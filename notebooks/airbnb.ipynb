#1st Method - DROPPING COLUMNS

```python
columns_to_drop = ['cancellation_policy','room type','Construction year','service fee', 'minimum nights',
       'number of reviews', 'last review', 'reviews per month',
       'review rate number', 'calculated host listings count',
       'availability 365', 'house_rules', 'license']
columns_to_keep = ['id', 'NAME', 'host id', 'host_identity_verified', 'host name',
       'neighbourhood group', 'neighbourhood', 'lat', 'long', 'country',
       'country code', 'instant_bookable']
```

```python
len(columns_to_drop)
```

```python
data[columns_to_keep]
```

```python
data.shape
```

```python
df = data[columns_to_keep]
df
```

```python
data.drop(columns=columns_to_drop, inplace=True)
```

```python
data.shape
```

```python
data
```

#2nd Method - RENAMING COLUMNS

```python
data.columns
```

```python
new_columns_name = []
for i in data.columns:
    new_columns_name.append(i.upper())
```

```python
new_dataframe_to_modify = data.copy()
```

```python
new_dataframe_to_modify.columns = new_columns_name
```

```python
data.columns
```

#3rd Method - DROPPING DUPLICATES

```python
data.duplicated().value_counts()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().value_counts()
```

#4th Method - REMOVE THE NaN VALUES

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

#5th Method - CLEANING INDIVIDUAL COLUMNS

```python
#data["NAME"] = data["NAME"].str.lower()
data = data.rename(columns={'NAME': 'name'})
```

```python
data.columns
```

```python
data.head()
```

```python
data["instant_bookable"] = data["instant_bookable"].astype(int)
```

```python
data.head()
```

```python
data["price"] = data["price"].str.replace("$","")
```

```python
data.head()
```

```python
data.reset_index(inplace=True)
```

```python
data.drop(columns = ["index"], inplace=True)
```

```python
data.head()
```

```python
data.drop(columns= ["level_0"], inplace=True)
```

```python
data.head()
```

#6th Method - DOWNLOAD DATA CLEANED

```python
data.to_csv("Airbnb_Open_Data_Cleaned.csv", index=False)
```
